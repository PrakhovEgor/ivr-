```python
base = "https://ru.wikipedia.org"
links = ["https://ru.wikipedia.org/w/index.php?title=%D0%9A%D0%B0%D1%82%D0%B5%D0%B3%D0%BE%D1%80%D0%B8%D1%8F:%D0%A0%D0%B0%D1%81%D1%82%D0%B5%D0%BD%D0%B8%D1%8F_%D0%BF%D0%BE_%D0%B0%D0%BB%D1%84%D0%B0%D0%B2%D0%B8%D1%82%D1%83&filefrom=%D0%90&subcatfrom=%D0%90&pageuntil=%D0%90%D0%BA%D0%B8+%28%D1%80%D0%B0%D1%81%D1%82%D0%B5%D0%BD%D0%B8%D0%B5%29#mw-pages"]
response = requests.get(links[0])
soup = BeautifulSoup(response.text, "html.parser")
while True:
    print(len(links))
    for i in soup.find_all("a"):
        if i.text == "Следующая страница":
            links.append(base + i["href"])
            response = requests.get(base + i["href"])
            soup = BeautifulSoup(response.text, "html.parser")
            break

    
```

```python
links
```

```python
links_plants = []
for j in links:
    response = requests.get(j)
    soup = BeautifulSoup(response.text, "html.parser")
    for i in soup.find("ul"):
        temp = i.find("a")
        if temp != -1:
            links_plants.append(base + temp["href"])
```

```python
len(links_plants)
```

```python
res = []
k = 1
for i in links_plants:
    try:
        response = requests.get(i)
        soup = BeautifulSoup(response.text, "html.parser")
        d = {}
        for j in soup.find("table", {"class": "infobox"}):
            d["name"] = j.find("th").text
            d["img_src"] = j.find("img")["src"]
            for div in j.find_all("div", {"class": "ts-Taxonomy-rang-row"}):
                if len(div["class"]) == 1:
                    try:
                        d[div.find("div", {"class": "ts-Taxonomy-rang-label"}).text] = div.find("a").text
                    except:
                        d[div.find("div", {"class": "ts-Taxonomy-rang-label"}).text] = div.find("b").text
                        
        description = ""
        for j in soup.find("p"):
            description += j.text
        d["description"] = description
        res.append(d)
    except:
        print('Error', i)
    print(k)
    k += 1
```

```python
import os
import requests
from PIL import Image
import io
from concurrent.futures import ThreadPoolExecutor

# Установите максимальный размер файла (в байтах)
max_file_size = 1024 * 1024  # Например, 1 МБ

# Создайте папку для сохранения изображений (если она не существует)
output_folder = '../static/plants'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

def download_image(i):
    url = i['img_src']
    url = "https:" + url
    url = url.replace("/thumb", "")
    url = url[:url.rfind("/")]
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        if len(response.content) > max_file_size:
            print("YEEEEE", i["name"])
            image = Image.open(io.BytesIO(response.content))
            image.thumbnail((image.width, 800))
            file_name = os.path.join(output_folder, i["name"] + ".jpg")
            image.save(file_name, format="JPEG")
        else:
            file_name = os.path.join(output_folder, i["name"] + ".jpg")
            with open(file_name, 'wb') as f:
                f.write(response.content)
    else:
        print(f'Ошибка при скачивании изображения {url}: статус код {response.status_code}')

# Используйте многопоточность для ускорения загрузки
with ThreadPoolExecutor(max_workers=6) as executor:
    executor.map(download_image, loaded_list)

print('Завершено скачивание изображений.')
```
